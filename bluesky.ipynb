```python
from atproto import Client, client_utils
import pandas as pd
handle = 'grogaldr.bsky.social'
password = 'vzwt-2b6b-gzgz-r3np'
max_page = 50
limit_per_page = 100
```

```python
keywords = [
    "AI",
    "DeepFake",
    "Artificial Intelligence",
    "ChatGPT",
    "OpenAI",
    "LLAMA",
    "Claude",
    "Anthropic",
    "Sora",
    "Sam Altman",
    "language model",
    "Machine Learning"
]

client = Client()
client.login(handle, password)

# 4. Function to search one keyword
def search_keyword(client, query, limit=100, max_pages=50):
    results = []
    cursor = None

    for _ in range(max_pages):
        response = client.app.bsky.feed.search_posts({
            "q": query,
            "limit": limit,
            "cursor": cursor
        })

        if not response.posts:
            break

        for post in response.posts:
            record = post.record

            results.append({
                "keyword": query,
                "uri": post.uri,
                "cid": post.cid,
                "author_handle": post.author.handle,
                "author_did": post.author.did,
                "text": getattr(record, "text", ""),
                "created_at": getattr(record, "createdAt", None),
                "indexed_at": getattr(post, "indexedAt", None),
                "reply_count": getattr(post, "replyCount", 0),
                "repost_count": getattr(post, "repostCount", 0),
                "like_count": getattr(post, "likeCount", 0),
                "quote_count": getattr(post, "quoteCount", 0),
                "languages": getattr(record, "langs", []),
                "embed": getattr(record, "embed", None),
            })

        cursor = response.cursor
        if not cursor:
            break

    return results


# 5. Collect for all keywords
all_results = []
for kw in keywords:
    print(f"Searching: {kw}")
    posts = search_keyword(client, kw)
    all_results.extend(posts)

# 6. Convert to DataFrame
df = pd.DataFrame(all_results)

# 7. Optional: remove duplicate posts (same URI may appear in multiple keyword searches)
df = df.drop_duplicates(subset=["uri"]).reset_index(drop=True)

# 8. Show preview
print(df.head())
print(f"\nTotal unique posts collected: {df.shape[0]}")

# 9. Save to CSV
df.to_csv("bluesky_keyword_posts.csv", index=False, encoding="utf-8")
```
