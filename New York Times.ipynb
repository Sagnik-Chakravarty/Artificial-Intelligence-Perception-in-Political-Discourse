```python
import os
import re
import pandas as pd

DIVIDER = r"_{30,}"   # matches long lines like ________________________

def clean(text):
    return re.sub(r"\s+", " ", text).strip() if text else None

def extract(pattern, text, flags=0):
    m = re.search(pattern, text, flags)
    return clean(m.group(1)) if m else None

def extract_block(pattern, text):
    m = re.search(pattern, text, flags=re.S)
    return clean(m.group(1)) if m else None

def parse_single_article(article_text):
    a = article_text

    data = {
        "title": extract(r"Title:\s*(.*)", a),
        "authors": extract(r"Author:\s*(.*)", a),
        "publication_title": extract(r"Publication title:\s*(.*)", a),
        "publication_date": extract(r"Publication date:\s*(.*)", a),
        "publication_year": extract(r"Publication year:\s*(.*)", a),
        "section": extract(r"Section:\s*(.*)", a),
        "pages": extract(r"Pages:\s*(.*)", a),
        "publisher": extract(r"Publisher:\s*(.*)", a),
        "place_of_publication": extract(r"Place of publication:\s*(.*)", a),
        "language": extract(r"Language of publication:\s*(.*)", a),
        "document_type": extract(r"Document type:\s*(.*)", a),
        "proquest_document_id": extract(r"ProQuest document ID:\s*(.*)", a),
        "url": extract(r"URL:\s*(.*)", a),
        "subjects": extract(r"Subject:\s*(.*)", a),
        "business_terms": extract(r"Business indexing term:\s*(.*)", a),
        "people": extract(r"People:\s*(.*)", a),
        "companies": extract(r"Company\s*/\s*organization:\s*(.*)", a),
        "locations": extract(r"Location:\s*(.*)", a),
        "abstract": extract_block(r"Abstract:\s*(.*?)(?=Links:|Full text:|\n[A-Z][a-z]+:)", a),
        "full_text": extract_block(r"Full text:\s*(.*?)(?=Subject:|Business indexing term:|Location:|URL:|Title:|$)", a),
    }

    # combined text
    data["text"] = clean(" ".join([t for t in [data["abstract"], data["full_text"]] if t]))

    return data


def process_folder(folder_path):
    rows = []

    txt_files = [
        f for f in os.listdir(folder_path)
        if f.endswith(".txt") and f.replace(".txt", "").isdigit()
    ]
    txt_files = sorted(txt_files, key=lambda x: int(x.replace(".txt", "")))

    for fname in txt_files:
        fpath = os.path.join(folder_path, fname)

        with open(fpath, "r", encoding="utf-8", errors="ignore") as f:
            raw = f.read()

        # split into ~500 articles
        chunks = re.split(DIVIDER, raw)

        for article in chunks:
            if len(article.strip()) < 100:  
                continue

            info = parse_single_article(article)
            info["file_name"] = fname
            rows.append(info)

    return pd.DataFrame(rows)




folder = "/Users/sagnikchakravarty/Desktop/AI Project/AI_Project/AI_Project/Data Collection/Newspaper/Jerusalem Post"
df = process_folder(folder)
df.head()
```

```python
import re
import pandas as pd

# ---------- 1. Read the RTF file ----------
# change the path if needed
path = "/Users/sagnikchakravarty/Desktop/AI Project/AI_Project/AI_Project/Data Collection/Newspaper/LA Times/500.RTF"

with open(path, encoding="utf-8", errors="ignore") as f:
    text = f.read()

# ---------- 2. Light RTF → plain text normalization (optional but helpful) ----------
# Convert RTF paragraph and tab markers into real newlines/tabs
text = re.sub(r'\\par[d]?', '\n', text)
text = re.sub(r'\\tab', '\t', text)

# (Very simple cleanup; if this strips too much, comment these two lines out)
# Remove some RTF control words / groups (header junk, fonts, etc.)
text = re.sub(r'{\\[^}]*}', '', text)          # simple control groups
text = re.sub(r'\\[A-Za-z]+\d* ?', '', text)   # control words like \fs24, \b, etc.

# ---------- 3. Trim off header before "1.  \l Bookmark_1 ..." ----------
m = re.search(r'\n1\.\s+Bookmark_?1', text)
if m:
    text = text[m.start():]

# ---------- 4. Split into individual records ----------
# Each record *in your file* starts with:
#  "N.  Bookmark_N ..."
raw_docs = re.split(r'\n(?=\d+\.\s+Bookmark_\d+)', text)

records = []

for doc in raw_docs:
    doc = doc.strip()
    if not doc:
        continue

    # ---- index (1, 2, 3, ...) ----
    m_idx = re.match(r'(\d+)\.', doc)
    idx = int(m_idx.group(1)) if m_idx else None

    # ---- headline (first quoted block) ----
    # e.g. "BUSINESS Meta invests in AI firm, recruits CEO ..."
    m_head = re.search(r'"(.*?)"', doc, flags=re.S)
    headline = m_head.group(1).strip() if m_head else None

    # ---- metadata line: date, source, section, author ----
    # In your text it looks like:
    # Jun 14, 2025\tLos Angeles Times\tMAIN NEWS; Business Desk; Part A; Pg. 1\tO'Brien writes for the Associated Press.
    m_meta = re.search(
        r'([A-Z][a-z]+ \d{1,2}, \d{4})\s+([^\t]+)\t([^\t]+)\t(.*)',
        doc
    )

    if m_meta:
        date, source, section_raw, author_raw = [g.strip() for g in m_meta.groups()]
    else:
        date = source = section_raw = author_raw = None

    records.append(
        {
            "index": idx,
            "headline": headline,
            "date": date,
            "source": source,
            "section_raw": section_raw,   # full string, e.g. "MAIN NEWS; Business Desk; Part A; Pg. 1"
            "author_raw": author_raw,     # often includes notes like "writes for Bloomberg."
            "raw_record": doc,            # full raw text for that hit (for anything you missed)
        }
    )

df = pd.DataFrame(records)

print(df.head())
print(df.columns)
```
